## 로지스틱 회귀

이번에 학습할 목표

1) K-최근접 이웃 모델<br>
K-최근접 이웃 모델을 활용해 이웃한 샘플의 클래스 비율을 구했지만 이는 갯수에 따라 항상 정해진 확률만을 출력해주는 문제가 발생했다.

2) 로지스틱 회귀<br>
로지스틱 회귀 모델은 회귀하는 **분류**모델이다. 선형회귀처럼 계산한 값을 그대로 출력하는 것이 아니라 0~1사이의 값으로 압축한다.

```python
import pandas as pd
import numpy as np

fish = pd.read_csv('http://bit.ly/fish_csv')
fish.head()
```

```python
print(pd.unique(fish['Species']))
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
print(fish_input[:5])

fish_target = fish['Species'].to_numpy()
print(fish_target[:5])
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

### K-최근접 이웃 분류기

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier(n_neighbors = 3)
kn.fit(train_scaled, train_target)
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))

# 사이킷런 모델에 타깃값을 전달하면 자동으로 순서를 매겨집니다.
# KNeighborsClassifier에서 정렬된 타깃값은 classes_ 속성에 저장되어 있다
print(kn.classes_)
```

### 다중분류

타깃 데이터에 2개 이상의 클래스가 포함된 문제의 경우 **다중 분류**라고 한다. <br>
**이진 분류**에서는 타깃값을 양성 클래스 1, 음성 클래스 0으로  지정하여 타깃 데이터를 만들었다.  <br>
다중 분류에서도 타깃값을 숫자로 바꾸어 입력할 수 있지만 사이킷런에서는 문자열 그대로 제공해줌<br>

```python
print(kn.predict(test_scaled[:5]))

# KNeighbors의 경우 샘플 X 주위의 가장 가까운 이웃 샘플 K개를 표시했습니다. 샘플 X의 확률을 종류별로 비율을 계산해서 내놓는다.
proba = kn.predict_proba(test_scaled[:5])
print(np.round(proba, decimals = 4))
print(kn.classes_)
```

```python
distances, indexes = kn.kneighbors(test_scaled[3:4])
print(train_target[indexes])
```

### 로지스틱 회귀

만약 KNeighbors에 해당하는 모델을 사용할 경우 내가 지정한 샘플에 대한 **확률값이 일정하다.**<br>
K = 3일 경우, 가능한 확률값은 0, 1/3, 2/3, 3/3이기 때문에 좋은 확률을 구하는 방법이 아니다.<br>

이에 따라 새로운 분류 모델을 생각해보는데, 이는 회귀이지만 분류 모델인 **로지스틱 회귀**이다. <br>

Z = ax1 + bx2 + cx3 + dx4 + ex5 + f 에 대한 선형방정식이 있다고 할 때, 우리가 원하는 것은 **Z의 값이 확률값**으로 나오기를 원한다.<br>
**0 <= Z <= 1**에 해당하는 값이 되기 위해서는 **시그모이드 함수**를 이용하여 앞선 선형방정식에 적용한다.




```python
## 시그모이드 활성함수를 numpy로 그리기
import numpy as np
import matplotlib.pyplot as plt
z = np.arange(-5, 5, 0.1)
phi = 1 / (1 + np.exp(-z))
plt.scatter(z, phi)
plt.xlabel('z')
plt.ylabel('phi')
plt.show()
```

```python
## 로지스틱 회귀는 LogisticRegression 클래스 이용
## 이진분류일 경우 0.5보다 크면 양성클래스 1, 0.5보다 작으면 음성클래스 0

char_arr = np.array(['A', 'B', 'C', 'D', 'E'])
print(char_arr[[True, False, True, False, False]])
```

```python
bream_smelt_indexes = (train_target == 'Bream') | (train_target == 'Smelt')
train_bream_smelt = train_scaled[bream_smelt_indexes]
target_bream_smelt = train_target[bream_smelt_indexes]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_bream_smelt, target_bream_smelt)
print(lr.predict(train_bream_smelt[:5]))


## 첫 번째 열: 음성클래스(0), 두 번째 열: 양성클래스(1)
## predict_proba를 통해 음성, 양성클래스에 대한 확률을 계산
print(lr.predict_proba(train_bream_smelt[:5]))
print(lr.classes_)
```

```python
## 로지스틱 회귀 모델이 학습한 방정식의 variance
print(lr.coef_, lr.intercept_)
```

```python
## 로지스틱 회귀 모델을 통과시킨 train_set값
## decision_function() 메서드의 경우 양성클래스에 대한 z값을 반환
decisions = lr.decision_function(train_bream_smelt[:5])
print(decisions)
```

```python
## sigmoid(f(train_set))에 통과시킨 값
from scipy.special import expit
print(expit(decisions))
```

### 로지스틱 회귀로 다중 분류

sklearn의 Logistic Regression 다중분류를 위해서는 해당 모델이 어떻게 동작하는지 알 필요가 있다.

1) max_iter : 학습 반복수 <br>
2) penalty : L1, L2 규제 방식 <br>
3) solver : 확률적 평균 경사 하강법 알고리즘 (saga추가)

```python
lr = LogisticRegression(C=20, max_iter=1000)
lr.fit(train_scaled, train_target)
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))


lrL = LogisticRegression(C=20, max_iter=1000, penalty="l1", solver="saga")
lrL.fit(train_scaled, train_target)
print(lrL.score(train_scaled, train_target))
print(lrL.score(test_scaled, test_target))
```

```python
print(lr.predict(test_scaled[:5]))

proba = lr.predict_proba(test_scaled[:5])
print(lr.classes_)
print(np.round(proba, decimals=3))
```

```python
## 다중분류의 경우 분류 갯수만큼 variance를 만든다
## 분류갯수: 7 -> W,b(7, 5) * X(5, 1) = Y(7, 1) > Y에 해당하는 값은 sigmioid X, softmax 함수를 사용하여 값이 변한다.
print(lr.coef_.shape, lr.intercept_.shape)
```

### 소프트맥스(Softmax)

**softmax**는 여러 개의 선형방정식의 출력값을 0~1사이로 압축하고 전체 합을 1이 되도록 만든다. 이는 다중분류에서 잘 쓰이는 **활성함수**이다. <br>



```python
from scipy.special import softmax

decision = lr.decision_function(test_scaled[:5])
print(np.round(decision, decimals=3))
print(np.round(softmax(decision, axis=1), decimals=3))

## softmax함수의 경우, 축을 지정하지 않으면 행과 열에 모두 적용하기 때문에 axis를 지정해야한다.
## axis = 1 (행) , axis = 0 (열)
```
