## 인공 신경망

```python
from tensorflow import keras 

(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
```

```python
print(train_input.shape, train_target.shape)
```

```python
print(test_input.shape, test_target.shape)
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(1, 10, figsize=(10, 10))
for i in range(10):
    axs[i].imshow(train_input[i], cmap='gray')
    axs[i].axis('off') # x, y 축에 대한 값을 OFF
plt.show()


print([train_target[i] for i in range(10)])
```

```python
import numpy as np
print(np.unique(train_target, return_counts=True))
```

```python
train_scaled = train_input / 255.0 ## 0-255의 train 값을 0-1사이의 값으로 변환
train_scaled = train_scaled.reshape(-1, 28*28) ## 2차원 배열 > 1차원 배열
print(train_scaled.shape) ## 784 feature로 이루어진 60000개의 샘플
```

### 교차검증(validation set)

- 보통은 train set으로 모델을 훈련, test set으로 모델을 검증
- 하지만 고정된 test set으로 모델의 성능을 검증하고 수정하는 일은, 머신러닝 모델을 test set에 과대적합한 모델이 될 수 있다.
- 이를 해결하고자 **교차검증**이 나왔는데, 이는 train set을 train set + validation set으로 분리하여, validation set으로 검증하는 방식이다

#### 장점
- 모든 데이터셋에 대하여 훈련할 수 잇다.
- 정확도 향상, 데이터 부족으로 인한 과소적합을 방지할 수 있다
- test set에 대한 과대적합을 막을 수 있다.
- 평가 결과에 따라 일반화된 모델을 만들 수 있다.

#### 단점
- iteration 횟수가 꽤 많다보니, 모델 훈련/평가 시간이 오래 걸린다.




```python
from sklearn.model_selection import cross_validate
from sklearn.linear_model import SGDClassifier

sc = SGDClassifier(loss='log', max_iter=5, random_state=42)
scores = cross_validate(sc, train_scaled, train_target, n_jobs=5)
print(np.mean(scores['test_score']))
```

```python
## cross_validate의 함수의 경우 fit_time, score_time, test_score 키를 가진 딕셔너리를 반환한다.
## 모델을 훈련하는 시간, 모델을 검증하는 시간, 모델 검증 score값
## 기본적으로 5-폴드 교차 검증을 수행

print(scores)
```

### 인공신경망

용어 정리
- 출력층 : z1~z10을 바탕으로 클래스를 예측하기 때문에 신경망의 최종 값
- 입력층 : x1~x784에 해당하는 입력값
- 뉴런(유닛) : 인공 신경망에서 z값을 계산하는 단위, 선형방정식으로 통한 계산



#### 딥러닝에서 교차검증

교차 검증을 잘 사용하지 않고 검증 세트를 별도로 덜어내어 사용하는 이유는?

1) 딥러닝 분야의 데이터셋은 충분히 크기 때문에 검증 점수가 안정적이다.<br>
2) 교차 검증의 경우 훈련 시간이 오래 걸리기 때문이다.

```python
from sklearn.model_selection import train_test_split

train_scaled, val_scaled, train_target, val_target = train_test_split(train_scaled, train_target, test_size=0.2, random_state=42)
print(train_scaled.shape, train_target.shape)
print(val_scaled.shape, val_target.shape)
```

### 완전 연결층(Fully Connected Layer)

양쪽의 뉴런이 모두 연결하고 있기 때문에 이런 층을 완전 연결층이라고 한다. 즉, Dense 클래스를 사용해 밀집층을 만들 수 있다.

### 활성화함수(Activation Function)
소프트맥스와 같이 뉴런이 선형 방정식이 계산 결과에 적용되는 함수를 **활성화함수**라고 한다.

```python
dense = keras.layers.Dense(10, activation='softmax', input_shape=(784,))
model = keras.Sequential(dense)
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics='accuracy')
```

### 원 - 핫 코드

분류 문제에서 타깃값을 해당 클래스만 1이고 나머지는 모두 0인 배열로 만드는 것이다.

```python
print(train_target[:10])
```

```python
model.fit(train_scaled, train_target, epochs=10)
```

```python
model.evaluate(val_scaled, val_target)
```
