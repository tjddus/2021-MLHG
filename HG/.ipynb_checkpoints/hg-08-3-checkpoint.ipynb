### 합성곱 신경망의 시각화

1) 가중치 시각화, 픽셀 강도에 따른 시각화

2) 합성곱 층에서 출력된 특성 맵 시각화

```python
from tensorflow import keras
model = keras.models.load_model('best-cnn-model.h5')
model.layers
```

```python
conv = model.layers[0]
print(conv.weights[0].shape, conv.weights[1].shape)
```

```python
conv_weights = conv.weights[0].numpy()
print(conv_weights.mean(), conv_weights.std())
```

```python
import matplotlib.pyplot as plt
plt.hist(conv_weights.reshape(-1, 1))
plt.xlabel('weights')
plt.ylabel('count')
plt.show()
```

```python
fig, axs = plt.subplots(2, 16, figsize=(15, 2))
for i in range(2):
    for j in range(16):
        axs[i, j].imshow(conv_weights[:, :, 0, i*16+j], vmin=-0.5, vmax=0.5)
        axs[i, j].axis('off')
plt.show()
```

```python
no_training_model = keras.Sequential()
no_training_model.add(keras.layers.Conv2D(32, kernel_size=3, activation='relu', padding='same', input_shape=(28, 28, 1)))
```

```python
no_training_model = no_training_model.layers[0]
print(no_training_model.weights[0].shape)
```

```python
no_training_weights = no_training_model.weights[0].numpy()
print(no_training_weights.mean(), no_training_weights.std())
```

```python
plt.hist(no_training_weights.reshape(-1, 1))
plt.xlabel('weight')
plt.ylabel('count')
plt.show()
```

```python
fig, axs = plt.subplots(2, 16, figsize=(15, 2))
for i in range(2):
    for j in range(16):
        axs[i, j].imshow(no_training_weights[:, :, 0, i*16+j], vmin=-0.5, vmax=0.5)
        axs[i, j].axis('off')
plt.show()
```

### 함수형 API

신경말 모델을 만들 때, keras의 Sequential 클래스를 이용했는데, 이 클래스는 층을 차례대로 쌓은 모델을 만듭니다. 입력이 2개, 출력이 2개인 경우가 있을 수 있기 때문에 앞선 클래스는 이용하기 어려울 수 있습니다. 이를 위해 **함수형 API**를 사용할 수 있습니다.

```python
dense1 = keras.layers.Dense(100, activation='sigmoid')
dense2 = keras.layers.Dense(10, activation='softmax')
```

```python
print(model.input)
```

```python
conv_acti = keras.Model(model.input, model.layers[0].output)
```

```python
(train_input, train_target), (test_input, test_target) = keras.datasets.fashion_mnist.load_data()
plt.imshow(train_input[0], cmap='gray_r')
plt.show()
```

```python
inputs = train_input[0].reshape(-1, 28, 28, 1) / 255.0
feature_maps = conv_acti.predict(inputs)
```

```python
print(feature_maps.shape)
```

```python
fig, axs = plt.subplots(4, 8, figsize=(15, 8))
for i in range(4):
    for j in range(8):
        axs[i, j].imshow(feature_maps[0, :, :, i * 8 + j])
        axs[i, j].axis('off')
plt.show()
```

```python
conv2_acti = keras.Model(model.input, model.layers[2].output)
```

```python
inputs = train_input[0:1].reshape(-1, 28, 28, 1) / 255.0
feature_maps = conv2_acti.predict(inputs)
```

```python
print(feature_maps.shape)
```

```python
fig, axs = plt.subplots(8, 8, figsize=(12, 12))
for i in range(8):
    for j in range(8):
        axs[i, j].imshow(feature_maps[0, :, :, i * 8 + j])
        axs[i, j].axis('off')
plt.show()

## 앞선 conv1_acti와 다르게 conv2_acti는 시각적으로 이해하기 어렵다. 
## 두번째 합성곱층의 필터 크기는 (3, 3, 32)인데 이는 첫번째 출력의 32 특성맵이 곱해져 다음 합성곱 층의 첫번째 특성 맵이 됩니다. 결국엔 (14, 14, 32)는 
## 어떤 부위를 감지하는지 직관적으로 이해하기 어렵다.

## 앞부분의 합성곱층은 시각적인 정보를 뒤쪽으로 갈수록 추상적인 정보를 학습한다.
```
