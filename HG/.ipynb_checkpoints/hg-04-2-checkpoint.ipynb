## 경사 하강법(Stochastic Gradient Descent)

1) **확률적 경사 하강법** <br>
훈련 세트에서 랜덤으로 추출된 한개에 데이터에 대해서 그래디언트를 계산한다. 전체 데이터를 사용하는 것이 아니라, 랜덤하게 추출한 일부 데이터를 사용한다는 점에서 **확률적**이라는 이름이 붙었다.

다만, 확률적 경사 하강법은 전체에서 하나의 샘플을 선택해 학습하는 것은 진폭이 크고 불안정하며, 속도도 매우 빠르다. 또한 데이터를 하나씩 처리하기 대문에 오차율도 매우 크고 GPU 성능을 모두 활용하지 못하는 단점이 있다. epoch 또한, 수십, 수백 번 이상 에포크를 수행해야 한다.

2) **미니배치 경사 하강법** <br>
확률적 경사 하강법에 대한 문제점을 보안하기 위해서 전체 학습데이터를 배치 사이즈로 나누어서 학습을 수행한다. 이는 Batch 보다 빠르고 SGD보다는 낮은 오차율을 가지기에 실전에서 자주 사용한다.

3) **배치 경사 하강법** <br>
모든 데이터를 한꺼번에 학습하는데 활용하기 때문에 부드럽게 수렴하나 샘플의 개수가 많아질수록 계산 시간이 많이 소요된다. 극단적인 경우는 데이터가 너무 많아서 한번에 전체 데이터를 모두 읽기도 어렵다.





### 손실함수 vs 비용함수

손실 함수는 샘플 하나에 대한 손실을 정의하고, 비용 함수는 훈련 세트에 있는 모든 샘플에 대한 손실 함수의 합을 말합니다. 

### 분류모델

#### 로지스틱 손실 함수(크로스엔트로피 손실 함수) 

이진 분류의 경우 4가지의 데이터가 있다고 가정했을 때,
예측(0.9, 0.3, 0.2, 0.8) == 타깃(1, 1, 0, 0)<br>
손실함수의 경우 예측과 타깃이 가까울 수록 작어야 하며, 예측과 타깃이 멀수록 커야한다.




### 회귀모델

#### 평균제곱오차

회귀모델의 경우, 타깃에서 예측을 뺀 값을 제곱한 다음 모든 샘플에 평균한 값입니다.

```python
import pandas as pd
fish = pd.read_csv("https://bit.ly/fish_csv")
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import SGDClassifier
sc = SGDClassifier(loss='log', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### 에포크와 과대/과소적합

확률적 경사 하강법 모델은 에포크 횟수에 따라 과대적합, 과소적합이 될 수 있습니다. 에포크 횟수가 적으면 과소적합된 모델일 가능성이 높고, 에포크 횟수가 많으면 과대적합 모델일 가능성이 높다.

```python
import numpy as np
sc = SGDClassifier(loss='log', random_state=42)
train_score=[]
test_score=[]
classes = np.unique(train_target)
```

```python
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

```python
sc = SGDClassifier(loss='log', max_iter=100, random_state=42, tol=None)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
sc = SGDClassifier(loss='hinge', max_iter=100, random_state=42, tol=None)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
